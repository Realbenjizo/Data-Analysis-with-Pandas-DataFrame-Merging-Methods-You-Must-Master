```python
import pandas as pd
```

```python
car_sales = pd.read_csv('car_sales_real.csv')
car_sales
```

```python
car_reviews = pd.read_csv('car_reviews_real.csv')
car_reviews
```

#### Inner Join

```python
car_reviews.merge(car_sales, left_on = 'Car_ID', right_on = 'Car_ID', 
                  how = 'inner') 
```

```python
car_reviews.merge(car_sales, left_on = 'Car_ID', right_on = 'Car_ID',
                  suffixes = ('_reviews', '_sales'),
                  how = 'inner') 
```

```python
car_reviews.merge(car_sales, left_on = ['Car_ID', 'Model'], 
                  right_on = ['Car_ID', 'Model'], how = 'inner') 
```

```python
pd.merge(car_reviews, car_sales, on=['Car_ID', 'Model'], how='inner')
```

#### Left Join

```python
pd.merge(car_reviews, car_sales, on=['Car_ID', 'Model'], how='left')
```

#### Right Join

```python
pd.merge(car_reviews, car_sales, on=['Car_ID', 'Model'], how='right')
```

#### Outer Join

```python
pd.merge(car_reviews, car_sales, on=['Car_ID', 'Model'], how='outer')
```
